```python
# Importing Scrapper libraries
import requests
from bs4 import BeautifulSoup
```

```python
# Fetch the page 
response = requests.get("https://www.iplt20.com/auction/2022")
```

```python
# Checking if it works or not 
print(response.status_code) # if it print 200 it means it is working 
```

```python
# Prase the HTML
soup =  BeautifulSoup(response.text, "lxml")
```

```python
# Findint the TOP BUYS 
heading = soup.find("h2", string="Top Buys")
# The table is the very next sibling element after that <h2>
table = heading.find_next("table")
print(table)  # should show the full table HTML
# print(heading)
```

```python
# finding h2 tags and their exact text
all_h2 = soup.find_all("h2")

for h2 in all_h2:
     print(repr(h2.get_text(strip=True))) #repr for hidden text 
```

```python
# find_all("tr") gets every row in the table
rows = table.find_all("tr")

for row in rows:
    cells = row.find_all("td")  # each cell in the row
    
    if len(cells) == 4:  # skip empty/header rows
        team   = cells[0].get_text(strip=True)
        player = cells[1].get_text(strip=True)
        role   = cells[2].get_text(strip=True)
        price  = cells[3].get_text(strip=True)
        
        print(f"{player} → {team} | {role} | {price}")
```
