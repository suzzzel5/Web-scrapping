```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops")
web
```

```python
soup = BeautifulSoup(web.content, "html.parser")
soup
```

```python
soup.prettify()
```

```python
names = soup.find_all("a", class_ = "title")
na = []
for name in names:
  print(name.text.strip())
  na.append(name.text.strip())
```

```python
na
```

```python
prices = soup.find_all("h4", class_="price")
pr = []
for price in prices:
  print(price.text.strip())
  pr.append(price.text.strip())
```

```python
pr
```

```python
descriptions = soup.find_all("p", class_="description")
desc = []
for description in descriptions:
  print(description.text.strip())
  desc.append(description.text.strip())
```

```python
desc
```

```python

ratings = soup.find_all("p",{'data-rating': True})
rate = []
for i in ratings:
  r = i.get("data-rating")
  rate.append(r)
  print(r)
```

```python
rate
```

```python
reviews = soup.find_all("span",{"itemprop":"reviewCount"})
review = []
for i in reviews:
  review.append(i.text.strip())
  print(i.text)
```

```python
review
```

```python
images = soup.find_all("img", class_="img-fluid")
image_links = []
for image in images:
    print(image.get("src"))
    image_links.append(image.get("src"))
```

```python
image_links
```

```python
data = {
    "Name":na,
    "Description":desc,
    "Rating":rate,
    "Reviews":review,
    "Images Links":image_links
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df.head()
```
